## Basic definitions



1. <font color='blue'>**Short rate:** </font> A short rate is the interest rate associated to an infinitesimally small unit of time and can be stochastic

1. <font color='blue'>**Money Saving Account $M(t)$:** </font> the value of a bank account at time $t$. We assume that $M(t_0) = 1$ and that it evolves following $dM_t = r_t M_t dt$ <br>
Hence a bank account with initial \\$1 at time 0 will grown to:<br><br>
$$\boxed{ M(t) = exp\left(\int_0^t r_sd_s\right)   }$$ <br>
Note that this is a stochastic variable !<br>

1. <font color='blue'>**Stochastic Discount Factor $D(t,T)$:** </font> The (stochastic) discount factor between date $t$ and date $T$ is given by
$$\boxed{D(t,T) = \dfrac{M(t)}{M(T)} = exp\left(-\int_t^T r_sd_s\right)  }$$ <br>

1. <font color='blue'>**Zero-coupon bond $P(t,T)$:** </font> A T-maturity zero-coupon bond is a contract that guarantees its holder the payment of one unit of currency at time T.
The contract value at time $t <T$ is denoted $P(t,T)$. Clearly $\forall T : P(T,T) = 1 $ <br><br>
Under the HJM framework we also have $$\boxed{ P(t,T) = \mathbb{E}\left[ D(t,T)\right] = \mathbb{E}\left[ exp\left(-\int_t^T r_sd_s\right)\right] }$$ <br>

1. <font color='blue'>**Continously-compounded spot interest rate $R(t,T)$:** </font> this is the continously compounded rate so that an investment worth $ P(t,T)$ at times $t$ earn 1 at time $T$. <br><br>
In other words we have $P(t,T) = e^{-R(t,T) (T-t)}$ from which we derive<br><br>
$$\boxed{R(t,T) =- \dfrac{\text{ln}P(t,T) }{T - t}  }$$ <br>

1. <font color='blue'>**Simply-compounded spot interest rate $L(t,T)$:** </font> this is the simply compounded rate so that an investment worth $ P(t,T)$ at times $t$ earn 1 at time $T$. <br><br>
In other words we have $P(t,T) = \dfrac{1}{1 + L(t,T)(T-t)} $ from which we derive<br><br>
$$\boxed{L(t,T) =- \dfrac{1 - P(t,T) }{P(t,T)(T - t)}  }$$ <br>

1. <font color='blue'>**Forward rate  $F(t;T_1,T_2)$:** </font> the forward interest rate at time $t$ for expiry $T_1$ and maturiy $T_2$ is defined <br><br>
    * Under continuous compounding form by $\boxed{F(t,T_1,T_2) = -\dfrac{\text{ln} P(t,T_2) - \text{ln} P(t,T_1) }{T_2 - T_1}} $ <br><br>
    
    * Under simply compounding form by  $\boxed{F(t,T_1,T_2) = \dfrac{1}{T-t}\left(\dfrac{P(t,T_1)}{P(t,T_2)} -1 \right)} $ <br>
    <br><br>

1. <font color='blue'>**Instantaneous forward interest rate  $f(t,T)$:** </font> the instantaneous forward interest rate at time $t$ for maturity $T$ is defined as <br><br>
$$\boxed{f(t,T):=  \lim_{T_1\to T}F(t;T_1,T) = -\dfrac{\partial }{\partial T}\text{ln }P(t,T)}$$ <br>

## Intro to Stochastic differential equations




This document provides a conceptual overview of the **Itô–Taylor expansion**—sometimes called the **stochastic Taylor expansion**—which generalizes the classical Taylor series to stochastic processes governed by stochastic differential equations (SDEs).

---

### Context: Stochastic Differential Equations

An SDE in **Itô form** typically looks like:

$$
\mathrm{d}X_t = a\bigl(X_t,t\bigr)\,\mathrm{d}t + b\bigl(X_t,t\bigr)\,\mathrm{d}W_t,
$$

where:

- $ X_t $ is the state variable (which can be multi-dimensional).  
- $ a(\cdot) $ is called the **drift** function.  
- $ b(\cdot) $ is called the **diffusion** (or volatility) function.  
- $ W_t $ is a standard Brownian motion (Wiener process).

The **Itô formula** (or **Itô’s lemma**) is a cornerstone result that gives a first-order expansion for a sufficiently smooth function $ f(X_t, t) $. However, when we want **higher-order expansions** for numerical methods or more refined approximations, we use the **Itô–Taylor expansion**.

---

### Deterministic vs. Stochastic Taylor Expansions

#### Deterministic Taylor Expansion

If you have a **deterministic** function $ f(x) $ and expand it about a point $ x_0 $, you get:

$$
f(x) = f(x_0) + f'(x_0)(x - x_0)
+ \frac{1}{2}f''(x_0)(x - x_0)^2 + \cdots
$$

This standard series relies on derivatives of $ f $ and powers of $ (x - x_0) $.

#### Stochastic (Itô–Taylor) Expansion

With **stochastic processes**, “increments” $ \mathrm{d}X_t $ involve both $ \mathrm{d}t $ (the drift part) and $ \mathrm{d}W_t $ (the diffusion part). Because $ \mathrm{d}W_t $ is random and of order $ \sqrt{\mathrm{d}t} $, the series includes extra terms to reflect **stochastic integrals** (like $ \int b\,\mathrm{d}W $) and **iterated integrals** (like $ \int \int \mathrm{d}W\,\mathrm{d}W $, $ \int \int \mathrm{d}W\,\mathrm{d}t $, etc.). These appear at higher orders in the expansion.

---

### Basic Idea of the Itô–Taylor Expansion

Consider a (possibly multi‐dimensional) function $ f(X_t) $. For simplicity, focus on one time step from $ t $ to $ t + \Delta t $. We want an expansion of:

$$
f\bigl(X_{t+\Delta t}\bigr) - f\bigl(X_t\bigr).
$$

Using the multi-dimensional Itô formula and keeping **terms up to a certain order** in $ \Delta t $ and $ \Delta W_t $ (where $ \Delta W_t = W_{t+\Delta t} - W_t $), the expansion takes the schematic form:

$$
f(X_{t+\Delta t})
\approx
f(X_t)
+ \sum_{\alpha} \Big[\text{(partial derivatives of } f) \times
\bigl(\text{powers of } \Delta t \text{ and integrals in } \Delta W_t\bigr)\Big],
$$

where the summation $ \sum_{\alpha} $ indicates all necessary partial‐derivative terms involving different combinations of $ \Delta t $ and $ \Delta W_t $.

#### Terms in the Expansion

1. **Drift Terms** (proportional to $ \Delta t $)  
   $$
   \int_t^{t+\Delta t} a(X_s, s)\,\mathrm{d}s \approx a(X_t, t)\,\Delta t.
   $$

2. **Diffusion Terms** (proportional to $ \Delta W_t $)  
   $$
   \int_t^{t+\Delta t} b(X_s, s)\,\mathrm{d}W_s \approx b(X_t, t)\,\Delta W_t.
   $$

3. **Higher‐Order Terms**  
   - Terms like $ (\Delta W_t)^2 $ are of order $ \Delta t $ in expectation (because $ \mathbb{E}[(\Delta W_t)^2] = \Delta t $).  
   - Mixed integrals like $ \int_t^{t+\Delta t} \int_t^{t+\Delta t} \dots \mathrm{d}W_s\,\mathrm{d}W_u $ can appear, and they are typically of smaller order but become essential in more precise approximations.  
   - Derivatives with respect to $ x $ combine with these increments. For a second‐order expansion, partial derivatives $ \frac{\partial^2 f}{\partial x^2} $ and cross derivatives appear.

Hence, **Itô–Taylor expansions** systematically collect all these terms up to a chosen order in $ \sqrt{\Delta t} $ (for diffusion) and $ \Delta t $ (for drift and second‐moments), producing an approximation for $ f(X_{t+\Delta t}) $.

---

### Example: One-Dimensional Second‐Order Expansion

For a single SDE

$$
\mathrm{d}X_t = a(X_t, t)\,\mathrm{d}t + b(X_t, t)\,\mathrm{d}W_t,
$$

and a sufficiently smooth $ f(x, t) $, the second‐order **Itô–Taylor** expansion over a small interval $ \Delta t $ often looks like:

$$
\begin{aligned}
f(X_{t+\Delta t}, t+\Delta t) 
- f(X_t, t)
&\approx 
\frac{\partial f}{\partial t}\Delta t
+ \frac{\partial f}{\partial x} \, b\,\Delta W_t
+ \frac{\partial f}{\partial x} \, a\,\Delta t
\\[6pt]
&\quad
+ \frac{1}{2} \frac{\partial^2 f}{\partial x^2}\,b^2\,\bigl(\Delta W_t^2 - \Delta t\bigr)
+ \cdots
\end{aligned}
$$

where:

- $ \Delta W_t = W_{t+\Delta t} - W_t $ is a Wiener increment, typically $ \mathcal{N}(0, \Delta t) $.  
- $ \Delta W_t^2 - \Delta t $ is included because $ \mathbb{E}[(\Delta W_t)^2] = \Delta t $, so $ (\Delta W_t)^2 - \Delta t $ is a “centered” second‐moment term.  
- Higher‐order terms would include cross terms (involving partial derivatives of $ a $ or $ b $) and possibly iterated stochastic integrals if one goes beyond second order.

This structure is used in **numerical schemes** (e.g., strong or weak approximations of SDEs) and to derive expansions for option pricing, filtering, etc.

---

### Why the Itô–Taylor Expansion Matters

1. **Numerical Simulation of SDEs**  
   - Higher‐order **strong** or **weak** numerical methods (e.g., Milstein scheme, stochastic Runge–Kutta) rely on Itô–Taylor expansions to achieve better accuracy in simulating sample paths.

2. **Approximation of Functionals**  
   - In derivative pricing, filtering theory, or risk analysis, we often need expansions of payoff or state‐dependent functions. The Itô–Taylor expansion helps systematically approximate those.

3. **Theoretical Insights**  
   - It shows how **stochastic integrals** add extra terms beyond the usual deterministic Taylor approach, reflecting the influence of Brownian motion’s volatility and correlations at higher orders.

---

### Key Takeaways

- The **Itô–Taylor expansion** generalizes the classical Taylor series to stochastic settings by including increments of Brownian motion and their higher‐order iterated integrals.  
- It is crucial in **stochastic calculus** for deriving higher‐order approximation schemes.  
- Each additional order introduces new derivative terms and higher‐order stochastic integrals (like $ \int \mathrm{d}W\cdot\mathrm{d}W $), capturing subtleties not seen in deterministic expansions.
