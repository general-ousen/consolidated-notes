# Affine models and Characteristic functions

## Characteristic functions

### Introductions

1. The Fourier trasnform is one of the most important transform in signal analysis. It transform one function from the time domain to the frequency domain

1. If $f(x)$ is a function then the fourier transform of $f(x)$ is defined as <br><br>
$$\phi(x) =  \int_{-\infty}^{+\infty}e^{iux}f(x)dx$$

1. The **Characteristic Function** of a random varaible $X$ denoted $\phi_X(u)$ is the fourier transform of its density function that is <br><br>
$$\boxed{\phi_X(u) := \int_{-\infty}^{+\infty}e^{iux}f(x)dx = \mathbb{E}\left[ e^{iuX}   \right]  }$$

1. In quant finance we often work with density functions of certain stochastic variables.

1. For some basic stochastic variables the density function is known in closed form

1. however for other few relevant stochastic processes in finance we do not know the coresponding density function.

1. The good news is that for some of these processes we can find the Characteristic Function which uniquely determines the distribution of $X$
    
1. Using fourier methods based on the characteristic function we can derive pricing methods that 
    * Are computationally very fast
    * Not restricted to Guaussian based models <br><br>

1. it helps find the moments via the simple relation 
$$\mathbb{E}\left[ X^k\right] = \dfrac{1}{i^k} \dfrac{\partial^k}{\partial u^k}\phi_X(u)\vert_{u=0}$$

1. for a log normal distributed variable X then we have $$\mathbb{E}\left[ X^k\right] = \phi_{logX}\left(-ik\right) = \int_{0}^{\infty}x^kf_X(x)dx$$
    
1. in math once can solve partial differential equation using what is called the "method of characteristics" this is used in quant finance.



### Characteristic function of the black scholes model

#### Feynman - Kac Theorem and the Blacksholes model



1. This theorem connects the solution of a specific class of partial differentials equations to an expectation.

1. Consider the Ito process  $$dS_t = \Theta(t,S_t)dt + \Psi(t, S_t)dW_t$$

1. Assume that $V(t,S_t)$ is the solution to the stochastic differential equation <br>  
$$ V_t +\Theta(t,S_t) V_s +\frac{1}{2} \Psi^2(t,S_t) V_{ss} - rV  = 0$$

1. With the boundary condition (pay-off function) $$V(T, S_T) = H(T,S_T)$$

1. The solution of the PDE is $$V(t,S_t) = \mathbb{E} \left[ H(X_T)e^{-\int_t^Trdu} \right]$$

1. if $r$ is constant we have $$V(t,S_t) = e^{-r(T-t)} \mathbb{E} \left[ H(T,S_T) \right]$$


#### Feynman - Kac Theorem and the Blacksholes model in log coordonate


1. The theorem also  holds in log coordonate

1. let set $X_t  = log S_t$ by ito lemma we have $$ dX_t = \left(r - \dfrac{1}{2}\sigma^2 \right) dt + \sigma dW_t^{\mathbb{Q}} $$ <br>

1. By this change of variable we can show that $X_t$ is solution to the following stochastic differential equation <br><br>
$$  V_t + r V_x + \dfrac{1}{2}\sigma^2 \left(  -V_x +V_{xx}  \right) - rV  = 0$$

1. With boundary / payoff $V(T,X) = H(T, X)$ 

1. The solution is by given by $$V(T, X) = e^{-r(T-t) } \mathbb{E}[H(T,X)]$$

#### Feynman - Kac Theorem and the Blacksholes model and characteristic function

1. Still working in the log coordonate assume that the final condition is now $$  H(T,X) = \phi_X(u; T,T) = e^{iuX_T}  $$

1. By Feyman Kac theorem the solution to the partil differencial equation is $$  \phi_X(u; t, T) = e^{-r (T-t)} \mathbb{E^Q}\left[  e^{iuX(T)} \right]  $$

1. That is the discounted chracteristic function of $X_t$ is a solution that is it satisfies<br><br>
$$  \phi_t + r \phi_x + \dfrac{1}{2}\sigma^2 \left(  -\phi_x +\phi_{xx}  \right) - r\phi  = 0$$

1. by Duffi-Pan-Singleton we have the solution is of the form $$  \phi_X(u; t, T) = e^{A(u,t,T) + B(u,t,T)X(t)} $$

1. With boundary coundition $\phi(u; T,T) = e^{iuX_T}$ by partial differenciation we have 
$$\begin{cases}
    \phi_t = \phi \times (A_t + X B_t )\\
    \phi_x = \phi \times B            \\
    \phi_{xx} = \phi \times B^2
\end{cases}$$

1. we replace in the partial differencial equation and find 
$$  \phi\left[ (A_t + X B_t ) +  r  \times B   + \dfrac{1}{2}\sigma^2 \left(  - B        + B^2  \right) -  r\right] = 0  $$

1. by replacing an solving we find we must have 
$$\begin{cases}
    B_t = 0 \\
    A_t = - rB - \dfrac{1}{2}\sigma^2 \left(  B^2- B     \right) +  r = -(r - \dfrac{1}{2}\sigma^2 )B
\end{cases}$$

1. using the boundary condition we find
$$\begin{cases}
    B(u,t,T) = iu \\
    A(u,t,T) = \left(r - \dfrac{1}{2}\sigma^2 \right)i u (T-t) - \dfrac{1}{2}\sigma^2 u^2 (T-t) - r(T-t)
\end{cases}$$

1. hence the obtained chracteristic function is $$\phi_X(u; t, T) = exp\left[ \left(r - \dfrac{1}{2}\sigma^2 \right)i u (T-t) - \dfrac{1}{2}\sigma^2 u^2 (T-t) - r(T-t) +  iu X \right]$$

##  Class of Affine Diffusion process

###  Definition

1. consider the system of SDE with uncorrelated brownian motion $dB_t$
$$d\mathbf{X}_t = \mathbf{\Theta}dt +  \color{magenta}{\mathbf{\Sigma}}d\mathbf{B}_t $$  <br>
This can be written as 
$$\begin{bmatrix}
          dX_1 \\
          \vdots \\
         dX_n \\
   \end{bmatrix}  = \begin{bmatrix}
                          \Theta_1 \\
                          \vdots \\
                         \Theta_n \\
                   \end{bmatrix} dt + 
\begin{bmatrix}
\Sigma_{1,1}& \cdots & \Sigma_{1,n} \\
\vdots & \ddots  & \vdots  \\
\Sigma_{n,1}& \cdots & \Sigma_{n,n} 
\end{bmatrix} 
\begin{bmatrix}
          dB_1 \\
          \vdots\\
         dB_n \\
   \end{bmatrix} 
$$ <br><br>

1. This model is in the affine diffusion class if we have all components are  linear combinaisons of the state variables $\mathbf{X}_t$. The components are
    * The drifts $\mathbf{\Theta} = a_0 + a_1 \mathbf{X_t}$ 
    * The variance terms  $\mathbf{\Sigma}^T\mathbf{\Sigma} =  (c_0)_{i,j} + (c_1)^T_{ij} \mathbf{X_t}$ 
    * The interest rate terms $r(\mathbf{X}_t) = r_0 +r_1^T \mathbf{X}(t)$
    
    

###  Discounted characteristic function
For an affine stochastic process $X(t)$ the discounted characteristic function is defined as : 

$$\phi(X(t), t,T,u) :=  \mathbb{E^\mathbb{Q}}\left[ e^{iuX_T}  e^{ -\int_{t}^{T}R(X_s)ds  }   \right] $$

with boundary condition : 

$$\phi(X(T), T,T,u) := e^{iuX_T}$$


###  Solution of the discounted characteristic function
Duffie, Pan and Singleton have shown that there exist a solution  to the discounted characteristic function for AD process of the form 

$$\phi(X(t), t,T,u) =  e^{A(u,t,T) + B(u,t,T)X(t)}$$

where the coefficients  $A(u,t,T)$ and $B(u,t,T)$ must satisfy the system of **_Riccati type ODE_**:

$$
 \begin{align}
 \frac{\partial{d}}{\partial\tau} A(u,\tau) &= -r_0 + B ^Ta_0+ \frac{1}{2}B ^Tc_0B ^T\\
 \frac{\partial{d}}{\partial\tau} B(u,\tau) &= -r_1 + a_1^T B + \frac{1}{2}B ^T c_1B\\
\end{align}
$$


###  Solution of the discounted characteristic function : black scholes example

For the geometrical brownian motion the process is given by $$dS_t = rS_tdt + \sigma S_tdW_t$$
This process is not affine because $\Psi(S_t)^2 = \sigma^2S_t^2$

Let's instead consider the stock log return and set $x(t) = lnS_t$ we have :

$$dlogS_t = dx_t = \left(r-\frac{1}{2}\sigma^2\right)dt + \sigma dW^{\mathbb{Q}}_t$$


This model is in the Affine Diffusion class with:
$$
 \begin{align}
 a_0 &= r-\frac{1}{2}\sigma^2  & a_1 = 0 \\
 c_0 &= \sigma^2  & c_1 = 0 \\
 r_0 &= r  & r_1 = 0 \\
\end{align}
$$


The Riccati ode are 

$$
\begin{equation}
    \begin{cases}
      \dfrac{\partial{d}}{\partial\tau} A(u,\tau) &= -r_0 + a_0B + \frac{1}{2}c_0B^2\\
 \dfrac{\partial{d}}{\partial\tau} B(u,\tau) &= -r_1 + a_1B + \frac{1}{2}c_1B^2\\
    \end{cases}       
\end{equation}
$$


by substitution that is 


$$
\begin{equation}
    \begin{cases}
      \dfrac{\partial{d}}{\partial\tau} A(u,\tau) &= -r + \left( r- \frac{1}{2}\sigma^2\right)B + \frac{1}{2}\sigma^2 B^2\\
      \dfrac{\partial{d}}{\partial\tau} B(u,\tau) &= 0\\
    \end{cases}       
\end{equation}
$$


By taking the boundary conditions $$e^{iuX} = e^{A(u,0) + B(u,0)X}$$ 

that is $A(u,0) = 0$ and $B(u,0) = iu$ we find the solutions 

$$
\begin{equation}
    \begin{cases}
      B(u,\tau) &= iu \\
     A(u\tau) &= \left[ -r +iu\left( r-\dfrac{1}{2}\sigma^2\right) - \dfrac{1}{2}\sigma^2u^2 \right]\tau \\
    \end{cases}       
\end{equation}
$$


hence the characteristic for the BM is 

$$\phi(u, \tau) = exp\left[ \left( r- \frac{1}{2}\sigma^2\right)iu\tau - \frac{1}{2}\sigma^2 u^2\tau -r\tau + iu\text{ log}S_0 \right]$$
