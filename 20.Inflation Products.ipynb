# Inflation Products

## Intro

1. It is common to model inflation in the foreign currency analogy.  According to this analogy 
    * real interest rate are the interest rate in the foreign/real economy
    * CPI is interpreted as the exchange rate between the domestic/nominal and foreign/real currencies <br><br>

1. There is a strong connection between nominal and real interest rate on the market and break-even inflation rate

1. Note you can have :
    * Nominal yield curves
    * Real yield curve
    * the diff is the Break even inflation rate

## Inflation Swap

1. An inflation swap has two leges : an _inflation leg_ and a _fixed leg_  with payoff at time $T$ <br><br>
$$V(T) = N \left(   \dfrac{I(T}{I(t_0)}  - (1 - K)^\alpha \right)$$
where 
    * $I(T)$ represent the CPI index at time $T$ 
    * $K$ is the fixed rate
 <br><br>
 
1. When workingwith inflation instruments it is important to realize that inflation data is typically delayed 

1. the forward inflation is given by $$I_F(t,T) = I(t)\dfrac{P_r(t,T)}{P_n(t,T)}$$
