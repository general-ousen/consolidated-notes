# Hybrid Models

## Introduction

1. Typycally when we compute xVA/PFE we compute it for a portfolio that has different asset classes or risk Factors for example a portfolio may be made of :
    * Stocks
    * Interest rate instruments
    * FX derivatives
    * inflation instruments 
    * etc...
    <br><br>
    
1. Hybrid models are used to define and simulate several correlated risk factors for different asset classes
  
1. The key model parameters needs to be calibrated to market instruments (european options)

1. Typically for hybrid models monte carlo simulations are used. Closed form solutions are quite rare

1. Hybrid models can also be used for valuing hybrid payoffs or exotic derivatives.

1. Hybrid models can be expressed by a system of correlated SDE like <br><br>
$$d\mathbf{X}_t = \Theta(t,\mathbf{X}_t)dt + \Psi(t, \mathbf{X}_t)d\mathbf{W}_t$$
Where :
    * $\mathbf{X}_t$ is a vector of risk factors from different asset classes
    * $d\mathbf{W}_t$ is a vector of brownian motion that are correlated 
    * $d\mathbf{\widetilde{W}}_t$ as the uncorrelated brownian motions
<br><br>
    
1. The speed of pricing European deivatives is crucial specially for calibrating the SDEs. This is why we focus to work with **characteristics functions** as working with these allows to use highly efficient pricing algorithm or European options such as 
    * The COS method
    * Fast Fourrier transform
