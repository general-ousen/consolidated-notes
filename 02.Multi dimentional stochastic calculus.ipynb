# Multi dimentional stochastic calculus

## Multi Dimentional Stochastic Differential Equation (SDE)

### Definitions

1. To talk of multi dimentional SDE we will start by studying brownnian motions $W(t)$ that are correlated and work toward defining them as independent brownian motions $B(t)$

1. To move from the correlated $W(t)$ and define them as uncorrelated  $B(t)$ via a technique called __Cholesky Decomposition__  of a correlation matrix is 

1. <font color='blue'>**Cholesky Decomposition:** </font> Each symetric positive definite matrix $\mathbf{C}$ has a unique factorization $\mathbf{L}$  which is a lower triangular matrix with positive diagonal entries that satisfies 
$$\mathbf{ C = LL}^T$$ <br>

1. We can then generate correlated brownian motion from uncorrelated brownian motion using <br><br>
$$\mathbf{W}(t) = \mathbf{L} \cdot \mathbf{B}(t)$$ 

1. For example
    * consider two brownian motions $\mathbf{W}_t := \begin{bmatrix}
           W_1(t) \\
          W_2(t) \\
         \end{bmatrix}$ 
     with correlation matrix  $\mathbf{C} := \begin{bmatrix}
          1 & \rho_{1,2} \\
         \rho_{1,2} & 1 \\
         \end{bmatrix}$ <br><br>
         
     
     * The Cholescky decomposition is $\mathbf{L} := \begin{bmatrix}
          1 &  0 \\
         \rho_{1,2} & \sqrt{1 - \rho_{1,2}^2} \\
         \end{bmatrix}$ <br><br>
         
     * We can generate these correlated brownian motion by matrix multiplication <br><br>
      $$\begin{bmatrix}
           W_1(t) \\
          W_2(t) \\
         \end{bmatrix} = \begin{bmatrix}
          1 &  0 \\
         \rho_{1,2} & \sqrt{1 - \rho_{1,2}^2} \\
         \end{bmatrix} 
             \cdot 
         \begin{bmatrix}
          B_1(t) \\
          B_2(t) \\
         \end{bmatrix} = 
         \begin{bmatrix}
           B_1(t) \\
         \rho_{1,2}B_1(t) +  B_2(t)  \sqrt{1 - \rho_{1,2}^2} \\
         \end{bmatrix} $$ <br>
         
      * that is <br>
      $$
                    \begin{align}
                    W_1(t) &= B_1(t) \\
                    W_2(t) &=\rho_{1,2}B_1(t) +  B_2(t)  \sqrt{1 - \rho_{1,2}^2} 
                    \end{align}
                    $$ <br><br>
                    

1. Consider the following system of correlated SDEs 
$$d\mathbf{X}_t = \mathbf{\Theta}(t, \mathbf{X}_t) dt + \mathbf{\Psi}(t, \mathbf{X}_t)d\mathbf{W}_t $$ <br><br>
Which is for simplicity   $$d\mathbf{X}_t = \mathbf{\Theta}dt + \mathbf{\Psi}d\mathbf{W}_t $$  <br><br>
This can be written as 
$$\begin{bmatrix}
          dX_1 \\
          \vdots \\
         dX_n \\
   \end{bmatrix}  = \begin{bmatrix}
                          \Theta_1 \\
                          \vdots \\
                         \Theta_n \\
                   \end{bmatrix} dt + 
\begin{bmatrix}
\Psi_{1,1}& \cdots & \Psi_{1,n} \\
\vdots & \ddots  & \vdots  \\
\Psi_{n,1}& \cdots & \Psi_{n,n} 
\end{bmatrix} 
\begin{bmatrix}
          dW_1 \\
          \vdots \\
         dW_n \\
   \end{bmatrix} 
$$ <br><br>

1. Note that in Quant finance typically the marix $\mathbf{\Psi}$ is a diagonal matrix 

1. Using Cholesky decomposion we have: 
$$d\mathbf{X}_t = \mathbf{\Theta}dt +  \color{magenta}{\mathbf{\Psi L}}d\mathbf{B}_t $$  <br>
Where $\mathbf{L}$ is the choleschy decomposition of the correlation matrix $\mathbf{C}$ between the $\mathbf{W}_t$.<br>
by changing varaiable can be rewriten as $$d\mathbf{X}_t = \mathbf{\Theta}dt +  \color{magenta}{\mathbf{\Sigma}}d\mathbf{B}_t $$  <br>
This can be written as 
$$\begin{bmatrix}
          dX_1 \\
          \vdots \\
         dX_n \\
   \end{bmatrix}  = \begin{bmatrix}
                          \Theta_1 \\
                          \vdots \\
                         \Theta_n \\
                   \end{bmatrix} dt + 
\begin{bmatrix}
\Sigma_{1,1}& \cdots & \Sigma_{1,n} \\
\vdots & \ddots  & \vdots  \\
\Sigma_{n,1}& \cdots & \Sigma_{n,n} 
\end{bmatrix} 
\begin{bmatrix}
          dB_1 \\
          \vdots \\
         dB_n \\
   \end{bmatrix} 
$$ <br><br>

1. <font color='blue'>**Ito lemma for vector processes:** </font>  consider we have a function $G(t, \mathbf{X}_t)$ then the dynamics of $G$ follows <br> <br> 
$$dG = G_t dt + \sum_{i=1}^n G_{x_i}dX_i + \frac{1}{2}\sum_{i,j=1}^n G_{x_i x_j }dX_idX_j$$
The compact expression above is equivalent to <br> <br> 
$$dG = \left( G_t + \sum_{i=1}^n \Theta_i G_{x_i} + \frac{1}{2} \sum_{i,j,k}^n \Sigma_{i,k}\Sigma_{j,k}   G_{x_i x_j }\right) dt + \sum_{i,j=1}^n \Sigma_{i,j}G_{x_i} dB_j(t)$$<br><br>





### Ito lemma 3D code

1. This code help finds the dynamics of $G(X_1, X_2, X_3)$ a function of 3 Ito proceses <br><br>
    $$ \begin{align}
                    dX_1 &= \Theta_1(t,X_1)dt + \Psi_1(t, X_1)dW_1 \\
                    dX_2 &= \Theta_2(t,X_2)dt + \Psi_2(t, X_1)dW_2 \\
                    dX_3 &= \Theta_3(t,X_3)dt + \Psi_3(t, X_1)dW_3 \\
       \end{align}
    $$
    
1. If the field **dependent_brownian** is set to true 
    - the results are shown with the correlated brownian motions $dW_s$
    - else Cholesky decomposition is done and results displayed as uncorrelated brownian motions $dB_s$ <BR><BR>
    
1. Takes as input :
    - The function $G$
    
    - The symbols for 
        - dW_1
        - dW_2
        - dW_3
    
    - The symbols for 
        - dB_1
        - dB_2
        - dB_3 
    
    - The correlation 
        - $ρ_{12} = corr(X_1, X_2)$
        - $ρ_{13} = corr(X_1, X_3)$
        - $ρ_{23} = corr(X_2, X_3)$
    
    - The Drift terms  
        - $\Theta_1$
        - $\Theta_2$
        - $\Theta_3$
    
    - The Diffusion terms  
        - $\Psi_1$
        - $\Psi_2$
        - $\Psi_3$
        

 





```python
from sympy import symbols, expand, factor, diff, E, Function, maximum, Derivative, collect, simplify , Matrix, log, apart, Equality
import sympy
def ito_lemma_3D (
G = Function('G')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
    
X_1 = symbols('X_1'),
X_2 = symbols('X_2'),
X_3 = symbols('X_3'),
    
dW_1 = symbols('dW_1'),
dW_2 = symbols('dW_2'),
dW_3 = symbols('dW_3'),   
    
dB_1 = symbols('dB_1'),
dB_2 = symbols('dB_2'),
dB_3 = symbols('dB_3'),  
    
ρ_12 = symbols('ρ_12'),
ρ_13 = symbols('ρ_13'),
ρ_23 = symbols('ρ_23'), 
       
Θ_1 = Function('Θ_1')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Θ_2 = Function('Θ_2')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Θ_3 = Function('Θ_3')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),

Ψ_1 = Function('Ψ_1')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Ψ_2 = Function('Ψ_2')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Ψ_3 = Function('Ψ_3')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),

factorize_it = False,
dependent_brownian = True
):


    dt, t = symbols(' dt,  t')

    Θ = Matrix([ 
          [Θ_1] ,
          [Θ_2] ,
          [Θ_3 ] 
    ])

    Ψ = Matrix([ 
          [ Ψ_1,  0      ,  0      ] ,
          [ 0      , Ψ_2 ,  0      ]  ,
          [ 0      ,   0     , Ψ_3 ] 
    ])

    dW = Matrix([ 
          [dW_1] ,
          [dW_2 ],  
          [dW_3 ]
    ])

    dB = Matrix([ 
          [dB_1] ,
          [dB_2 ] ,
          [dB_3 ]  
    ])

    C = Matrix([ 
          [1, ρ_12, ρ_13] ,
          [ρ_12, 1, ρ_23] ,
          [ρ_13 ,ρ_23, 1]
    ])

    L = C.cholesky(hermitian=False)
    Σ = Ψ*L
    
    print("The system of SDE is : ")
    if dependent_brownian :
        dX = Θ * dt + Ψ*dW
        display(dX)
        dX_1 = dX[0]
        dX_2 = dX[1]
        dX_3 = dX[2]
        
        
    else:
        dX = Θ * dt + Σ*dB
        display(dX)
        dX_1 = dX[0]
        dX_2 = dX[1]
        dX_3 = dX[2]



    


    G_t = diff(G, t)

    G_x1 = diff(G, X_1) 
    G_x2 = diff(G, X_2) 
    G_x3 = diff(G, X_3) 

    G_x1_x1 = diff(G_x1, X_1) 
    G_x2_x2 = diff(G_x2, X_2)
    G_x3_x3 = diff(G_x3, X_3)

    G_x1_x2 = diff(G_x1, X_2) 
    G_x1_x3 = diff(G_x1, X_3) 

    G_x2_x1 = diff(G_x2, X_1) 
    G_x2_x3 = diff(G_x2, X_3) 

    G_x3_x1 = diff(G_x3, X_1) 
    G_x3_x2 = diff(G_x3, X_2) 


    dG = G_t*dt + G_x1 * dX_1  + G_x2 * dX_2 +  G_x3 * dX_3  +  1/2* (G_x1_x1 * dX_1 * dX_1 + 
                                                                   G_x2_x2 * dX_2 * dX_2 + 
                                                                   G_x3_x3 * dX_3 * dX_3 + 

                                                                   G_x1_x2 * dX_1 * dX_2 +
                                                                   G_x2_x1 * dX_2 * dX_1 +  

                                                                   G_x1_x3 * dX_1 * dX_3 +
                                                                   G_x3_x1 * dX_3 * dX_1 +

                                                                   G_x2_x3 * dX_2 * dX_3 +
                                                                   G_x3_x2 * dX_3 * dX_2                                                       
                                                                  )  

    dG = (
    expand (dG)
    .subs(dB_1**2, dt)
    .subs(dB_2**2, dt)
    .subs(dB_3**2, dt)
    .subs(dB_1*dt, 0)
    .subs(dB_2*dt, 0)
    .subs(dB_3*dt, 0)
    .subs(dB_1*dB_2, 0)
    .subs(dB_1*dB_3, 0)
    .subs(dB_2*dB_3, 0)
    .subs(dt**2,0)
    .subs(dW_1**2, dt)
    .subs(dW_2**2, dt)
    .subs(dW_3**2, dt)
    .subs(dW_1*dt, 0)
    .subs(dW_2*dt, 0)
    .subs(dW_3*dt, 0)
    .subs(dW_1*dW_2, ρ_12*dt)
    .subs(dW_1*dW_3, ρ_13*dt)
    .subs(dW_2*dW_3, ρ_23*dt)
    .subs(dt**2,0)
    .subs(G, symbols('G'))
    )
    
    return collect(dG , (dt,dB_1,dB_2, dB_3, dW_1,dW_2, dW_3, G ) )  if factorize_it else dG
```

### Ito lemma example

In the example below we have the system of equations 
$$\begin{align}
dS_1(t) & = r  S_1(t) dt + \sigma_1  dW_1(t) \\
dS_2(t) & = r  S_2(t)  dt + \sigma_2  dW_2(t) \\
dM(t) & = r  M(t)  dt  \\
\end{align}$$ 

we use Ito lemma 3D to find the dynamic of $$G(t, S_1, S_2, M_t) = \dfrac{ S_2} {S_1}$$

```python
dt, t, r = symbols(' dt,  t, r')

X_1 = S_1 = symbols('S_1')
X_2 = S_2 = symbols('S_2')
X_3 = M_t = symbols('M_t')

G = S_2 / S_1 

dW_1 = symbols('dW_1')
dW_2 = symbols('dW_2')
dW_3 = symbols('dW_3')

dB_1 = symbols('dB_1')
dB_2 = symbols('dB_2')
dB_3 = symbols('dB_3')

ρ_12 = symbols('ρ')
ρ_13 = 0
ρ_23 = 0

Θ_1 = r * S_1 
Θ_2 = r * S_2
Θ_3 = r * M_t

Ψ_1 = symbols('σ_1') * S_1 
Ψ_2 = symbols('σ_2') * S_2 
Ψ_3 = 0

dG = ito_lemma_3D ( G = G,
    
                    X_1 = X_1,
                    X_2 = X_2,
                    X_3 = X_3,
                    
                    dW_1 = dW_1,
                    dW_2 = dW_2,
                    dW_3 = dW_3,   
                    dB_1 = dB_1,
                    dB_2 = dB_2,
                    dB_3 = dB_3,  

                    ρ_12 = ρ_12,
                    ρ_13 = ρ_13,
                    ρ_23 = ρ_23, 

                    Θ_1 = Θ_1,
                    Θ_2 = Θ_2,
                    Θ_3 = Θ_3,

                    Ψ_1 = Ψ_1,
                    Ψ_2 = Ψ_2,
                    Ψ_3 = Ψ_3,

                    factorize_it = True, dependent_brownian=False
)

print(f"\n We have the dynamic of G(t, X_1, X_2, X_3) follows\n")
dG#.simplify()
```

## Change of measure and randon Nykodyn

### Definition

1. <font color='blue'>**Radon-Nikodym Derivative** </font>  Consider we have  
    * Numeraire $X_1$ and its generated measure $\mathbb{1}$
    
    * Numeraire $X_2$ and its generated measure $\mathbb{2}$
    
    * The Radon-Nikodym derivative with define measure $\mathbb{2}$ is given by <br><br>
    
    $$\lambda^{\mathbb{2}}_{\mathbb{1}}(T) := \dfrac{d \mathbb{2}}{d \mathbb{1}} = \dfrac{X_2(T)}{X_2(t_0)}\dfrac{X_1(t_0)}{X_1(T)}$$ 
    <br><br>
    
    
1. <font color='blue'>**Girsanov theorem** </font> : Consider the stochastic process under two measure $\mathbb{1}$ and $\mathbb{2}$ <br><br>
$$\begin{align}
dX_t &= \Theta^{\mathbb{1}}(t, X_t) dt + \Psi(t, X_t) dW^{\mathbb{1}}_t  \\
dX_t &= \Theta^{\mathbb{2}}(t, X_t) dt + \Psi(t, X_t) dW^{\mathbb{2}}_t  \\
\end{align}$$
The link between the two measure is 
$$dW^{\mathbb{1}}_t = dW^{\mathbb{2}}_t + \Psi_\lambda $$ <br>
where $\Psi_\lambda = \dfrac{\Theta^{\mathbb{2}} - \Theta^{\mathbb{1}} }{ \Psi} $

1. Note that because $\lambda^{\mathbb{2}}_{\mathbb{1}}(T)$ is divided by the numeraire $X_1(T)$, it is a martingale under $\mathbb{1}$ hence we can write <br><br>
$$d \lambda^{\mathbb{2}}_{\mathbb{1}}(t) =  \lambda^{\mathbb{2}}_{\mathbb{1}}(t) \Psi_\lambda dW^{\mathbb{1}}_t $$ <br><br>




### Change of measure matrix 

1. Let us define two equivalent measure $\mathbb{1}$ and $\mathbb{2}$ <br><br>
    
    1. Numeraire $X_1(t)$ defined by the process :   $$dX_1 = \Theta_1( t, X_1)dt + \Psi_1(t, X_1)  dW^{\mathbb{1}}_1(t) $$ <br><br>
    
    1. Numeraire $X_2(t)$ defined by the process :   $$dX_2 = \Theta_2( t, X_2)dt + \Psi_2(t, X_2)  dW^{\mathbb{1}}_2(t) $$ <br><br>
    
    1. With correlation between the process : $$dW^{\mathbb{1}}_1(t)\times  dW^{\mathbb{1}}_2(t) = \rho_{12} dt$$ <br><br>
    
    
1. Consider and Asset $X_3$ with dynamic defined by $$dX_3 = \Theta^{\mathbb{1}}_3( t, X_3)dt + \Psi_3(t, X_3)  dW^{\mathbb{1}}_3(t) $$ <br><br>

1. We can simplify the notation as  $$d\begin{bmatrix}
         X_1 \\
         X_2 \\
         X_3 \\
   \end{bmatrix}  = \begin{bmatrix}
                          \Theta_1 \\
                          \Theta_2 \\
                         \Theta_3\\
                   \end{bmatrix} dt + 
\begin{bmatrix}
\Psi_1 & 0 & 0 \\
0 & \Psi_2  & 0  \\
0 & 0 & \Psi_3
\end{bmatrix} \mathbf{L} dB^{\mathbb{1}}(t) \\   
$$ <br><br>

1. Equivalently that is writing 
$$\begin{align}
dX_1 & = \Theta_1( t, X_1)dt + \Sigma_1(t, X_1)  dB^{\mathbb{1}}_1(t) \\
dX_2 & = \Theta_2( t, X_2)dt + \Sigma_2(t, X_2)  dB^{\mathbb{1}}_2(t) \\
dX_3 & = \Theta_3( t, X_3)dt + \Sigma_3(t, X_3)  dB^{\mathbb{1}}_3(t) \\
\end{align}$$ 
<br><br>
Where $\Sigma_1(\cdot), \Sigma_2(\cdot)$ and $\Sigma_3(\cdot)$ are row vectors from multiplying the covaraiance matrix with $\mathbf{L}$ that is :
<br><br>
$$\begin{align}
\Sigma_1 & = \begin{bmatrix} \Psi_1 \mathbf{L}_{1,1} & 0 & 0 \end{bmatrix}\\
\Sigma_2 & = \begin{bmatrix} \Psi_2 \mathbf{L}_{2,1} & \Psi_2 \mathbf{L}_{2,2} & 0 \end{bmatrix}\\
\Sigma_3 & = \begin{bmatrix} \Psi_3 \mathbf{L}_{3,1} & \Psi_3 \mathbf{L}_{3,2} & \Psi_3 \mathbf{L}_{3,3} \end{bmatrix}\\
\end{align}$$ <br><br>

1. The asset $X_3$ with dynamic under $\mathbb{2}$ has the form  $$dX_3 = \Theta^{\mathbb{2}}_3( t, X_3)dt + \Sigma_3(t, X_3)  dB^{\mathbb{2}}_3(t) $$ <br><br>

1. We want to find the drift term $\Theta^{\mathbb{2}}_3( t, X_3)$ as we know that the volatility term does not change from measure to measure<br><br>

1. From     
    * we also know that  $$\lambda^{\mathbb{2}}_{\mathbb{1}}(T) = \dfrac{X_1(t)}{X_1(T)}\dfrac{X_2(T)}{X_2(t)}$$ 
    
    * by applying ito lemma with we have 
    $$d \lambda^{\mathbb{2}}_{\mathbb{1}}(t) = \lambda^{\mathbb{2}}_{\mathbb{1}}(t) \left( \dfrac{\Sigma_2}{X_2} - \dfrac{\Sigma_1}{X_1}  \right) dB^{\mathbb{1}}_t 
    $$
    
     * The girsanov theorem we know that the radon nykodym derivative is a martingale that is  : $$d \lambda^{\mathbb{2}}_{\mathbb{1}}(t) =  \lambda^{\mathbb{2}}_{\mathbb{1}}(t) \Psi_\lambda d\mathbf{B}^{\mathbb{2}} $$
<br><br>
1. By Identification we have $$d \lambda^{\mathbb{2}}_{\mathbb{1}}(t) = \lambda^{\mathbb{2}}_{\mathbb{1}}(t) \left( \dfrac{\Sigma_2}{X_2} - \dfrac{\Sigma_1}{X_1}  \right) dB^{\mathbb{1}}_t =  \lambda^{\mathbb{2}}_{\mathbb{1}}(t) \Sigma_\lambda  d\mathbf{B}^{\mathbb{1}}_t$$
That is 
$$ \dfrac{\Sigma_2}{X_2} - \dfrac{\Sigma_1}{X_1} = \dfrac{\Theta^{\mathbb{2}} - \Theta{\mathbb{1}} }{ \Sigma}$$
Hence we have 
$$\Theta^{\mathbb{2}} = \Theta^{\mathbb{1}} + \Sigma \left(  \dfrac{\Sigma_2}{X_2} - \dfrac{\Sigma_1}{X_1} \right)$$

### Change of measure CODE

The code below allows to execute the measure change from measure 1 to measure 2 that is 

If we start with correlated Brownian motions and dynamics defined in $\mathbb{1}$
<br><br>
$$\begin{align}
dX_1 & = \Theta_1( t, X_1)dt + \Psi_1(t, X_1)  dW^{\mathbb{1}}_1(t) \\
dX_2 & = \Theta_2( t, X_2)dt + \Psi_2(t, X_2)  dW^{\mathbb{1}}_2(t) \\
dX_3 & = \Theta_3( t, X_3)dt + \Psi_3(t, X_3)  dW^{\mathbb{1}}_3(t) \\
\end{align}$$ 

We end with uncorrelated Brownian motions and dynamics defined in $\mathbb{2}$
<br><br>
$$\begin{align}
dX_1 & = \Theta_1^{\mathbb{2}}( t, X_1)dt + \Sigma_1(t, X_1)  dB^{\mathbb{2}}_1(t) \\
dX_2 & = \Theta_2^{\mathbb{2}}( t, X_2)dt + \Sigma_2(t, X_2)  dB^{\mathbb{2}}_2(t) \\
dX_3 & = \Theta_3^{\mathbb{2}}( t, X_3)dt + \Sigma_3(t, X_3)  dB^{\mathbb{2}}_3(t) \\
\end{align}$$ 

```python


def change_measure_from_X1_to_X2 (
    
X_1 = symbols('X_1'),
X_2 = symbols('X_2'),
X_3 = symbols('X_3'),

dX_1 = symbols('dX_1'),
dX_2 = symbols('dX_2'),
dX_3 = symbols('dX_3'),
    
dW_1 = symbols('dW_1'),
dW_2 = symbols('dW_2'),
dW_3 = symbols('dW_3'),   
    
dB_1 = symbols('dB_1'),
dB_2 = symbols('dB_2'),
dB_3 = symbols('dB_3'),  
    
ρ_12 = symbols('ρ_12'),
ρ_13 = symbols('ρ_13'),
ρ_23 = symbols('ρ_23'), 
       
Θ_1 = Function('Θ_1')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Θ_2 = Function('Θ_2')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Θ_3 = Function('Θ_3')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),

Ψ_1 = Function('Ψ_1')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Ψ_2 = Function('Ψ_2')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3')),
Ψ_3 = Function('Ψ_3')(symbols('t'), symbols('X_1'), symbols('X_2'), symbols('X_3'))

):
    dt, t = symbols(' dt,  t')
    
    dX = Matrix([ 
          [dX_1] ,
          [dX_2] ,
          [dX_3] 
    ])

    Θ = Matrix([ 
          [Θ_1] ,
          [Θ_2] ,
          [Θ_3 ] 
    ])

    Ψ = Matrix([ 
          [ Ψ_1,  0      ,  0      ] ,
          [ 0      , Ψ_2 ,  0      ]  ,
          [ 0      ,   0     , Ψ_3 ] 
    ])

    dW = Matrix([ 
          [dW_1] ,
          [dW_2 ],  
          [dW_3 ]
    ])

    dB = Matrix([ 
          [dB_1] ,
          [dB_2 ] ,
          [dB_3 ]  
    ])

    C = Matrix([ 
          [1, ρ_12, ρ_13] ,
          [ρ_12, 1, ρ_23] ,
          [ρ_13 ,ρ_23, 1]
    ])

    L = C.cholesky(hermitian=False)
    
    Σ = Ψ*L

    Σ_1 =  Σ.row(0)
    Σ_2 =  Σ.row(1)
    Σ_3 =  Σ.row(2)

    A = (  Σ_2 / X_2 - Σ_1 / X_1 )
    B = Σ_3 * A.T 

    Θ_new = Θ  + Σ * (  Σ_2 / X_2 - Σ_1 / X_1 ).T

    dX_new =  Θ_new *dt +  Σ * dB

    dX_new = (
    simplify(expand (dX_new)
    .subs(dB_1**2, dt)
    .subs(dB_2**2, dt)
    .subs(dB_3**2, dt)
    .subs(dB_1*dt, 0)
    .subs(dB_2*dt, 0)
    .subs(dB_3*dt, 0)
    .subs(dB_1*dB_2,0)
    .subs(dB_1*dB_3,0)
    .subs(dB_2*dB_3, 0)
    .subs(dt**2,0)
    .subs(dW_1**2, dt)
    .subs(dW_2**2, dt)
    .subs(dW_3**2, dt)
    .subs(dW_1*dt, 0)
    .subs(dW_2*dt, 0)
    .subs(dW_3*dt, 0)
    .subs(dW_1*dW_2, ρ_12*dt)
    .subs(dW_1*dW_3, ρ_13*dt)
    .subs(dW_2*dW_3, ρ_23*dt)
    ))
    
    expression = Equality(dX, dX_new)
    display(expression)

    return  dX_new
    
    
    
    
```

### Change of measure Example
