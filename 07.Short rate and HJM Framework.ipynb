## Affine term-structure model 

1. Affine term structure models is an interest rate models where the continously compounded rate  $R(t,T)$ is an affine function of the short rate that is: <br><br>
$$R(t,T) = \alpha(t,T) + \beta(t,T) r(t)$$
<br> Where $\alpha$ and $\beta$ are deterministic function of time

1. This relation is always true when the zero coupon bond can be writen as: <br><br>
$$P(t,T) = A(t,T)e^{-B(t,T)r(t)}$$
<br>

1. Forward rate vol in affine models:  <br><br>
$$\sigma_f(t,T) = \dfrac{\partial B(t,T)}{\partial T}\sigma\left(t, r(t) \right)$$ 
<br> where 
    * $\sigma_f(t,T):$ is the volatility of the instantaneous forward rate
    * $\sigma(t,r(t)):$ is the diffusion term of the short rate dynamic

## Equilibrium vs no-arbitrage models

1. An endogenous/no-arbitrage model, is a model designed to be consistent with today's term structure of interest rates. Here today's term structure of interest rate is an input. <br> 

1. An exogenous/equilibrium model, is a model where the term structure of interest rate is an output <br> 

1. some popular exogenous/equilibrium model are 
    * The vacicek model

## The HJM framework

### Overview

1. The Heath-Jarrow-Morton framework represents a class of models that are derived by modeling the dynamics of instantaneous forward rates (instead of instantaneous short rates)

1. HJM provides an explicit relation between volatility of the instantaneous forward rates and arbitrage free drift. By setting different format of the volatilities we may get different models

1. With HJM the model fits directly the current yield curve and leave some free parameters that can be used to fit other instruments

1. Both the standard short-rate and Libor Market models can be derived in the HJM framework

1. The HJM models are non-Markovian so only a few number of models have a closed-form solution

### General HJM Dynamics

1. Assume that instantaneous forward rate under $\mathbb{P}$ is driven by <br><br>
$$df(t,T) = \alpha^{\mathbb{P}}(t,T)dt + \sigma(t,T)dW^{\mathbb{P}}_t$$<br>
Then the instantaneous forward rates under  $\mathbb{Q}$ is driven by <br><br>
$$df(t,T) = \alpha^{\mathbb{Q}}(t,T)dt + \sigma(t,T)dW^{\mathbb{Q}}_t$$<br>
where $$\boxed{\alpha^{\mathbb{Q}}(t,T) = \sigma(t,T)\int_t^T\sigma(t,s)ds}$$

1. By using the relation $f(t,t) =r(t)$ we can find the dynamic of the short rate dynamic as <br><br>
$$\boxed{  dr(t) = \left[  \frac{\partial}{\partial t}f(0,t) + \alpha^{\mathbb{Q}}(t,t)  + \int_0^t\frac{\partial}{\partial t}\alpha^{\mathbb{Q}}(s,t)ds   +  \int_0^t\frac{\partial}{\partial t}\sigma(s,t)dW^{\mathbb{Q}}_s\right]dt + \sigma(t,t) dW^{\mathbb{Q}}_t  }$$

1. By setting different parametrization for instantenous forward rates vol $\sigma(t,T)$ we get different HJM model 
    * Example of Ho-Lee: $\text{ }$ $\sigma(t,T) = \sigma$ 
    * Example of HW1F: $\text{ }$ $\sigma(t,T) = \sigma e^{-\lambda(T - t)}$
<br><br>

1. The dynamics of zero coupon bonds under the risk neutral measure is given by the following lemma :<br><br>
$$ \boxed{ dP(t,T) = r(t)P(t,T)dt - P(t,T) \left( \int_t^T\sigma(z,T)dz \right)dW_t^{\mathbb{Q} }   } $$<br>
