# The Heston Hullwhite hybrid model

1. Under the Heston hullwite mode we have the following SDE under $\mathbb{Q}$ <br><br>
$$
 \begin{align}
dS_t &= rS_tdt +\sqrt{v_t} S_tdW_s^{\mathbb{Q}} \\
dv_t &= \kappa (\zeta - v_t)dt + \gamma\sqrt{v_t}dW_v^{\mathbb{Q}} \\
dr_t &= \lambda (\theta_t -r_t)dt + \eta dW_r
\end{align}
$$<br>
with correlation between the brownian motions $\rho_{s,v}$  $\rho_{r,v}$ and $\rho_{s,r}$

1. This is not in the affine class even after trying to use log coordinate hence it is not possible to directly derive the charateristic function. 

1. To remedy this use the strategy used is to approximate the term $\eta\rho_{s,r}\sqrt{v_t} \approx \eta\rho_{s,r}\mathbb{E}\left[\sqrt{v_t}\right]$ which makes the model to become affine

1. We now the distribution of a cir process is non central Chi-squared hence we can find in closed form $\mathbb{E}\left[\sqrt{v_t}\right]$.

1.  However this exact close form formula is  expensive to compute. An alternative is to compute the approximation via <br><br>$$ \mathbb{E}\left[  \sqrt{v_t} \right] \approx \sqrt{\mathbb{E}[v_t]  - \frac{\mathbb{Var}[v_t]}{4 \mathbb{E}[v_t]}} := \mathcal{E}(t)$$

1. Worth noting that this approximation works best when the feller condition is true

1. an even better approximation that helps makes computaion faster and eases finding chracteristic function is <br><br>
$$\mathbb{E}\left[  \sqrt{v_t} \right] \approx  a + b e^{-c t}: \tilde{\mathcal{E}(t)}$$
Where <br> 
$
 \begin{align}
&a = \sqrt{\zeta - \dfrac{\gamma^2}{8\kappa}},\\ &b = \sqrt{v(0)} - a, \\ &c= -log\left[ b^{-1} (\mathcal{E}(1) - a)  \right]
\end{align}
$

1. Using the appoximation above we call this model the **H1-HW model** and its characteristic function <br><br>
