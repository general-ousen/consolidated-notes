# HW1F model calibration

## Theory and Pricing see (Brigo & Mercurio Chapter 1, 2,3)

1. <font color='blue'>**Short rate dynamics under $\mathbb{Q}$:**</font>  The HW model is given by <br><br>
$$dr(t) = \left[ \vartheta(t) -ar(t)\right]dt + \sigma dW^{\mathbb{Q}}(t) $$
<br>
The distribution of the short rate $r(t)$ conditional on $\mathscr{F}_s$ <br><br> 
$$ r(t)|\mathscr{F}_s \sim \mathscr{N}\left(  r(s)e^{-a(t-s)}  + \alpha(t) - \alpha(s)e^{-a (t-s)},   \dfrac{\sigma^2}{2a}\left[ 1 - e^{-2a(t-s)}\right] \right)   $$
<br>
Where <br>
$$\begin{align}
\vartheta(t) & = -\dfrac{\partial f^M(0,T)}{\partial T} + af^M(0,t) + \dfrac{\sigma^2}{2a}(1 - e^{-2at}) \\
\alpha(t)  &= f^M(0,t) + \dfrac{\sigma^2}{2a^2}(1 - e^{-2at})^2
\end{align}
$$
<br>
<br>
1. <font color='blue'>**Affine properties:**</font> <br><br>
$$P(t,T) = A(t,T)e^{-B(t,T)r(t)}$$
<br>
$
\begin{align}
B(t,T) &= \dfrac{1-e^{-a(T-t)}}{a} \\
A(t,T) &=   \dfrac{P^M(0,T)}{P^M(0,t)} \text{exp}\left[  B(t,T)f^M(0,t) - \dfrac{\sigma^2}{4a}\left(  1 - e^{-2at} \right)B(t,T)^2 \right]
\end{align}
$
<br><br>

1. <font color='blue'>**Zero Bond Option pricing:**</font> The price at time $t$ of The Zero Bond Option price can be found analytically as <br><br>
$$\text{ZBO}(t, T, S, X) =  \omega\left[  P(t,S)\phi(\omega h) -  X P(t,T)\Phi\left( \omega \left(h-\sigma_p\right) \right) \right]  $$
where
    * $\omega$ : 1 for a call -1 for a put
    * $t$ : todays date
    * $T$ : Maturity date of the option
    * $S$ : Maturity date of the Zero coupon bond
    * $\Phi$ : standard normal cdf
    * $\sigma_p = \sigma \sqrt{    \dfrac{(1 - e^{-2a(T-t)} )}{2a}    }B(T,S) $
    * $h = \dfrac{1}{\sigma_p} \text{ln}\dfrac{P(t,S)}{P(t,T)X}+ 0.5\sigma_p$
<br>
<br>

1. <font color='blue'>**Cap/Floor Pricing:**</font> Caps/Floors can be seen as a portfolio of zero-bonds-options <br><br>
$$
\begin{align}
\text{Cap}(t, \mathcal{T}, N, X) &= N\sum_{i=1}^{n} (1 + X\tau_i)  \text{ZBP}\left(t, t_{i-1}, t_i, \dfrac{1}{1 + X\tau_i}\right) \\
\text{Flr}(t, \mathcal{T}, N, X) &= N\sum_{i=1}^{n} (1 + X\tau_i)  \text{ZBC}\left(t, t_{i-1}, t_i, \dfrac{1}{1 + X\tau_i}\right)
\end{align}
$$
Where
    * $t$ is todays date
    * $D = \{d_1, d_2,...,d_n\}$ the set of payments dates on the cap/floor
    * $\mathcal{T} = {t_0, t_1, t_2,....t_n}$ the set of corresponding time in year such that $t_i = d_i - t$
    * $t_0$ is the first reset date
    * $\tau_i = d_i - d_{i-1}$
    * $\sigma_p^i = \sigma\sqrt{  \dfrac{1 - e^{-2a(t_{i-1} - t )} }{ 2a}  }B(t_{i-1} , t_i)$
    * $ h_i =  \dfrac{1}{\sigma_p^i} \text{ln}\dfrac{P(t,t_i)(1 + X \tau_i)}{P(t, t_{i-1})} + 0.5 \sigma_p^i$
<br>
<br>

1. <font color='blue'>**Coupon Bearing Bond Option CBO:**</font>  priced explicily via Jmshidian's decomposition<br><br>
$$
\begin{align}
\text{CBO}(t, T,\mathcal{T}, c, X) &= \sum_{i=1}^{n} c_i \text{ZBO}\left(t, T, T_i,X_i\right) \\
\end{align}
$$
Where
    * $t$ is todays date
    * $\mathcal{T} = {T_1, T_2,..., T_n}$ the set of payment time
    * $c:$ cashflow array
    * $r^*$ the value of the spot rate at time $T$ for which the coupon bearing bond $CB$ is equal to the strike price
    * $X_i$ the value of a pure discount bond maturing at $T_i$ when the spot rate is $r^*$
<br><br>    
    
1. <font color='blue'>**Swaption Pricing:**</font>  a swaption can be seen as an option on a coupon bearing bond<br><br>
$$
\begin{align}
\text{Swaption}^R(t, T,\mathcal{T}, N, X) &= \sum_{i=1}^{n} X\tau_i \text{ZBC}\left(t, T, t_i,X_i\right) \\
\text{Swaption}^P(t, T,\mathcal{T}, N, X) &= \sum_{i=1}^{n} X\tau_i \text{ZBP}\left(t, T, t_i,X_i\right) \\
\end{align}
$$
Where
    * $t$ is todays date
    * $\mathcal{T} = {T_1, T_2,..., T_n}$ the set of payment time
    * $c_i = X\tau_i$  with $c_n = 1 + X\tau_i$
    * $r^*$ such that  $\sum_i^{n}c_i A(T,t_i) e^{-B(T,t_i)r^*} = 1$
    * $X_i$ the value of a pure discount bond maturing at $T_i$ when the spot rate is $r^*$
    


## Details on Time dependent mean reversion and volatility methodology

1. The strategy used here is outline by the paper from Mizuho securities here https://people.kth.se/~aaurell/Teaching/SF2975_HT17/calibration-hull-white.pdf

1. We consider a parameter time grid $\{t_i\}$ where $i= 0,...,N-1$ where $N$ is the number of points on the grid.

1. we model the mean reversion rate $a(t)$ and $\sigma(t)$ as piece wise constant between the time grids where we have <br><br>
$$
\begin{equation}
a(t) =
    \begin{cases}
       a_i, & t \in [t_i, t_{i+1}) & \text{for } i = 0,...N-2 \\
    a_{N-1}, & t > t_{N-1}\\
    \end{cases}    \\  
\sigma(t) =
    \begin{cases}
       \sigma_i, & t \in [t_i, t_{i+1}) & \text{for } i = 0,...N-2 \\
    \sigma_{N-1}, & t > t_{N-1}\\
    \end{cases}
\end{equation}
$$
1. We choose a logistic function for the form of the mean reversions that is : $a_i = A_0 + \dfrac{A_1 - A_0}{1 + e^{A_2 (A_3 - t_i)}}$
    * $A_0$ : represent the short term value
    * $A_1$ : represent the long term value
    * $A_2$ : represent the the slope / how fast the mean reversion transitions
    * $A_3$ : represent the point in time where the transition occurs
    * <kbd>
    
1. For the volatility we consider a grid of volatility vs time $\{ T_k, \Sigma_k \}$ and set $\sigma_i = CubicSpline(t_i, \{T_k\}, \{\Sigma_k\}) $ where :
    * $T_k$ represents the time points
    * $\Sigma_k$ represent volatility parameters

```python
def zero_coupon_to_yield(T, z):
    return -log(z) / T * 100



def P(t, T):
    return P0T_market(T) / P0T_market(t)


def τ(t, T, convention='years'):
    if convention == 'years':
        return (T - t)


def R(t, T):
    return - log(P(t, T)) / τ(t, T)


def L(t, T):
    return (1 - P(t, T)) / (τ(t, T) * P(t, T))


def F(t, T, S):
    τ = S-T
    return 1 / τ * (P(t, T) / P(t, S) - 1)


def f(t, T):
    ϵ = 1e-5
    return -(log(P(t, T + ϵ)) - log(P(t, T - ϵ))) / (2 * ϵ)


def FRA(t, T, S, N, K, fra_type='R'):
    '''
    t : Valuation date
    T : Reset date
    S : Payment date
    N : Notional
    K : Strike
    fra_type : if R then receive fixed rate else Pay
    '''
    τ = S-T
    α = 1 if fra_type == 'R' else -1
    return N * P(t, S) * τ * (K - F(t, T, S)) * α


def IRS(t, T, τ, N, K, irs_type='R'):
    '''
    t : valuation date
    T : array of payment dates
    τ : year fraction floating index
    N : Notional
    K : Swap Rate
    '''
    α = 1 if irs_type == 'R' else -1

    return sum(FRA(t, T - τ, T, N, K, fra_type=irs_type))


def CB(t, T, c):
    '''
    t : is the valuation date
    T : is an array of time where the cashflows are received
    c : is an array of coupon payments
    '''
    return sum(P(t, T) * c)


def FRN(t, T, τ, N):
    '''
    t : todays date
    T : payment dates
    '''
    return IRS(t, T, τ, N, 0, irs_type='P') + N * P(t, T[-1])


def S_α_β(t, T, τ):
    T_α = T[0] - τ
    T_β = T[-1]
    S = (P(t, T_α) - P(t, T_β)) / sum(τ * P(t, T))
    return S


def Bl(K, F, ν, ω):
    def d1(K, F, ν): return (log(F / K) + 0.5 * ν ** 2) / ν

    def d2(K, F, ν): return (log(F / K) - 0.5 * ν ** 2) / ν

    return F * ω * norm.cdf(ω * d1(K, F, ν)) - K * ω * norm.cdf(ω * d2(K, F, ν))


def cap_black(t, T, τ, N, K, σ_α_β):
    ν_i = σ_α_β * sqrt(T - τ)
    return N * sum(P(t, T) * τ * Bl(K, F(t, T - τ, T), ν_i, 1))


def floor_black(t, T, τ, N, K, σ_α_β):
    ν_i = σ_α_β * sqrt(T - τ)
    return N * sum(P(t, T) * τ * Bl(K, F(t, T - τ, T), ν_i, -1))


def Swaption_black(t, T, τ, N, K, σ_α_β, swaption_type='R'):
    '''
    t : valuation date
    T : array of payment dates
    τ : year fraction floating index
    N : Notional
    K : Swap Rate
    '''
    ω = 1 if swaption_type == 'P' else -1
    par = S_α_β(t, T, τ)
    T_α = T[0] - τ

    return N * Bl(K, par, σ_α_β * sqrt(T_α), ω)


def ZBO_vasicek(t, T, S, X, option_type='C', **vasicek_params):
    '''
    See book page 60 eq 3.10

    t : todays date
    T : maturity date of the option
    S : Maturity date of the the zero coupon bond
    option_type : 'C' for call else put

    Vasicek parameters:
        k : mean reversion parameter of the
        θ : long run average
        σ : short rate volatility rate
        r_t : short rate value
    '''
    k = vasicek_params['k']
    θ = vasicek_params['θ']
    σ = vasicek_params['σ']
    ω = 1 if option_type.upper() == 'C' else -1

    def B(t, T): return 1 / k * (1 - exp(-k * (T - t)))

    def A(t, T): return exp((θ - σ ** 2 / (2 * k)) * (B(t, T) - T + t) - σ ** 2 / (4 * k) * B(t, T) ** 2)

    def P(t, T): return A(t, T) * exp(-B(t, T) * r_t)

    σ_p = σ * sqrt((1 - exp(-2 * k * (T - t))) / (2 * k)) * B(T, S)
    h = 1 / σ_p * log(P(t, S) / (P(t, T) * X)) + 0.5 * σ_p
    φ = norm.cdf

    ZBO = ω * (P(t, S) * φ(ω * h) - X * P(t, T) * φ(ω * (h - σ_p)))
    return ZBO


def ZBO_cir(t, T, S, X, option_type='C', **cir_params):
    pass


def A_hullwhite(t, T, a, σ):
    B = lambda t, T: B_hullwhite(t, T, a)
    return P(0, T) / P(0, t) * exp(B(t, T) * f(0, t) - σ ** 2 / (4 * a) * (1 - exp(-2 * a * t)) * B(t, T) ** 2)


def B_hullwhite(t, T, a):
    return 1 / a * (1 - exp(-a * (T - t)))


def P_hullwhite(t, T, a, σ, r_t):
    '''
    Computes the price of zero coupon bond P(t,T) using hullwhite model
    t : valuation date
    T : Payment date of ZCB
    HW parameters:
        a : mean reversion parameter of the
        σ : short rate volatility rate
        r_t : short rate value r(t)
    '''

    def B(t, T): return B_hullwhite(t, T, a)

    def A(t, T):  return A_hullwhite(t, T, a, σ)

    return A(t, T) * exp(- B(t, T) * r_t)


def F_HW(t, T, S,a, σ, r_t):
    def P(t,T) : return P_hullwhite(t, T, a, σ, r_t)
    τ = S-T
    return 1 / τ * (P(t, T) / P(t, S) - 1)

def FRA_HW(t, T, S, N, K,a, σ, r_t, fra_type='R'):
    '''
    t : Valuation date
    T : Reset date
    S : Payment date
    N : Notional
    K : Strike
    fra_type : if R then receive fixed rate else Pay
    '''
    def P(t,T) : return P_hullwhite(t, T, a, σ, r_t) 
    def F(t,T,S) : return F_HW(t, T,S, a, σ, r_t) 
    τ = T -S
    I = np.where(S >t, 1,0)
    # print(I)

    α = 1 if fra_type == 'R' else -1
    return N * P(t, S) * τ * (K - F(t, T, S)) * α * I


def IRS_HW(t, T, τ, N, K,a, σ, r_t, irs_type='R'):
    '''
    t : valuation date
    T : array of payment dates
    τ : year fraction floating index
    N : Notional
    K : Swap Rate
    '''
    α = 1 if irs_type == 'R' else -1
    
    def FRA(t, T, S, N, K) : return FRA_HW(t, T, S, N, K,a, σ, r_t, fra_type=irs_type)

    return sum(FRA(t, T - τ, T, N, K))

def CB_hullwhite(t, T, c, a, σ, r_t):
    '''
    t : is the valuation date
    T : is an array of time where the cashflows are received
    c : is an array of coupon payments
    '''

    P = lambda t, T: P_hullwhite(t, T, a, σ, r_t)
    return sum(P(t, T) * c)


def ZBO_HW(t, T, S, X, a, σ, r_t, option_type='C'):
    '''
    See book page 76 eq 3.4

    t : todays date
    T : maturity date of the option
    S : Maturity date of the the zero coupon bond
    option_type : 'C' for call else put
    HW parameters:
        a : mean reversion parameter of the
        σ : short rate volatility rate
    '''

    ω = 1 if option_type.upper() == 'C' else -1

    P = lambda t, T: P_hullwhite(t, T, a, σ, r_t)

    def B(t, T): return B_hullwhite(t, T, a)

    σ_p = σ * sqrt((1 - exp(-2 * a * (T - t))) / (2 * a)) * B(T, S)
    h = 1 / σ_p * log(P(t, S) / (P(t, T) * X)) + 0.5 * σ_p
    φ = norm.cdf

    ZBO = ω * (P(t, S) * φ(ω * h) - X * P(t, T) * φ(ω * (h - σ_p)))
    return ZBO


def cap_hullwhite(t, T, τ, N, X, a, σ, r_t):
    '''
    See book page 76 eq 3.4
    t : todays date
    T : Array of payment times
    τ : time frac between payment date and reset date
    HW parameters:
        a : mean reversion parameter of the
        σ : short rate volatility rate
    '''
    ZBP = lambda t, t_1, t_2, X: ZBO_HW(t, t_1, t_2, X, a, σ, r_t, option_type='P')
    return N * sum((1 + X * τ) * ZBP(t, T - τ, T, 1 / (1 + X * τ)))


def floor_hullwhite(t, T, τ, N, X, a, σ, r_t):
    '''
    See book page 76 eq 3.4
    t : todays date
    T : Array of payment times
    τ : time frac between payment date and reset date
    HW parameters:
        a : mean reversion parameter of the
        σ : short rate volatility rate
    '''
    ZBC = lambda t, t_1, t_2, X: ZBO_HW(t, t_1, t_2, X, a, σ, r_t, option_type='C')
    return N * sum((1 + X * τ) * ZBC(t, T - τ, T, 1 / (1 + X * τ)))


def CBO_hullwhite(t, T, Γ, c, X, a, σ, r_t, option_type='C'):
    '''
    Compute the value of an option on a coupon bearing bond under the HW1F
    t : todays date
    T : expiry date of the option
    Γ : array of coupon payment date of the bond
    c : array of cashflows of the coupon bond
    '''

    # first find r* such that the value of the bond is equal to the strike price
    f = lambda r: CB_hullwhite(T, Γ, c, a, σ, r) - X
    r_star = newton(f, 0.0)

    # find X_i the value at time T of a ZCB maturing at time Γ_i
    X_i = P_hullwhite(T, Γ, a, σ, r_star)

    #
    CBO = sum(c * ZBO_HW(t, T, Γ, X_i, a, σ, r_t, option_type=option_type))
    return CBO


def Swaption_hullwhite(t, T, Γ, τ, X, N, a, σ, r_t, swaption_type='R'):
    '''
    Compute the value of Swaption under the HW1F
    t : todays date
    T : expiry date of the option
    Γ : array of  payment date of the underlyng swap
    τ : time frac between payment date and reset date
    '''
    # compute the array c
    c = ones_like(Γ) * X * τ
    c[-1] = (1 + X * τ)

    # first find r* such that the value of the bond is equal to the strike price
    f = lambda r: CB_hullwhite(T, Γ, c, a, σ, r) - 1

    try:
        r_star = newton(f, 0)
    except RuntimeError:
        try:
            r_star = brentq(f, -1, 1)
        except Exception:
            r_star = brentq(f, -1000, 1000)

    # find X_i the value at time T of a ZCB maturing at time Γ_i
    X_i = P_hullwhite(T, Γ, a, σ, r_star)

    ZBP = lambda t, t_1, t_2, X: ZBO_HW(t, t_1, t_2, X, a, σ, r_t, option_type='P')
    ZBC = lambda t, t_1, t_2, X: ZBO_HW(t, t_1, t_2, X, a, σ, r_t, option_type='C')

    #
    if swaption_type == 'R':
        return N * sum(c * ZBC(t, T, Γ, X_i))
    else:
        return N * sum(c * ZBP(t, T, Γ, X_i))


def to_time(x):
    m = re.split('^([0-9]+)', x)

    if m[2] == 'Mo':
        return float(m[1]) / 12

    if m[2] == 'Yr':
        return float(m[1])


def compute_S_market(T_1, T_2, τ, σ_black):
    Γ = arange(T_1, T_1 + T_2, τ) + τ
    X_atm = S_α_β(0,Γ, τ)
    S_market = Swaption_black(t=0, T=Γ, τ=τ, N=1, K=X_atm, σ_α_β=σ_black)

    return S_market


def compute_S_hw(T_1, T_2, τ, a, σ):
    Γ = arange(T_1, T_1 + T_2, τ) + τ
    X_atm = S_α_β(0, Γ, τ)
    S_hw = Swaption_hullwhite(t=0, T=T_1, Γ=Γ, τ=τ, X=X_atm, N=1, a=a, σ=σ, r_t=f(0, 0), swaption_type='R')

    return S_hw


def hw1f_mle(params):
    a = params[0]
    b = params[1]
   
    T_1 = df.T_1
    T_2 = df.T_2
    τ = df.τ
    S_market = df.S_market
    σ_market = df.Black_vol
    S_hw = zeros_like(S_market)
    σ_hw = zeros_like(S_market)
    
    for i, val in enumerate(S_market):
        S_hw[i] = compute_S_hw(T_1[i], T_2[i], τ[i], a, b)
        #σ_hw[i] = compute_black_σ(T_1[i], T_2[i], τ[i], S_hw[i])*100

    squared_err = (S_hw/S_market -1) ** 2
    mean_squared_err = squared_err.mean()
    RMSE = sqrt(mean_squared_err)
    return RMSE


def compute_black_σ(T_1, T_2, τ, S):
    Γ = arange(T_1, T_1 + T_2, τ) + τ
    X_atm = S_α_β(0, Γ, τ)
    f = lambda x: Swaption_black(t=0, T=Γ, τ=τ, N=1, K=X_atm, σ_α_β=x) - S

    try:
        σ_hat = newton(f, 0)
    except RuntimeError:
        try:
            σ_hat = brentq(f, -1, 1)
        except Exception:
            σ_hat = brentq(f, -1000, 1000)

    return σ_hat
```

### generate Short rate paths

```python
import numpy as np
M = 1_000 # number of paths
N = 30*250 # number of steps
T  = 30
ϵ = 0.001
θ   = lambda t : (f(0,t + ϵ) - f(0,t-ϵ))/(2.0*ϵ) + λ*f(0,t)+  σ**2 / (2.0*λ)*(1.0-np.exp(-2.0*λ*t))

Z = np.random.normal(0.0,1.0,[M,N])
W = np.zeros([M, N+1])
R = np.zeros([M, N+1])
M_t = np.zeros([M, N+1])
M_t[:,0] = 1
R[:,0]=f(0,0)
t = np.zeros([N+1])
dt = T / float(N)


for i in range(0,N):
    W[:,i+1] = W[:, i] + Z[:,i]*sqrt(dt)
    dW = W[:,i+1] - W[:,i]
    R[:, i+1] =  R[:, i] +  (θ(t[i]) - λ*R[:, i] )*dt + σ*dW 
    M_t[:, i+1] = M_t[:,i] * np.exp( (R[:, i+1] + R[:, i])/2.0 * dt)
    t[i+1] = t[i]+ dt


fig = plt.figure(figsize=(14,8)) # create a figure where we will put our plots
plt.plot(t, np.transpose(R[0:25,:]), linestyle='-')
plt.grid()
plt.xlabel("time")
plt.ylabel("r(t)");
plt.title('Hull White 1F simulation', fontsize = 24);

```

## Example PFE Calc for interest rate swaps

```python
import numpy as np
M = 1_000 # number of paths
N = 1000 # number of steps
T  = 10 # maturity of the swap
ϵ = 0.001
Notional = 100

θ   = lambda t : (f(0,t + ϵ) - f(0,t-ϵ))/(2.0*ϵ) + λ*f(0,t)+  σ**2 / (2.0*λ)*(1.0-np.exp(-2.0*λ*t))

Z = np.random.normal(0.0,1.0,[M,N])
W = np.zeros([M, N+1])
R = np.zeros([M, N+1])
M_t = np.zeros([M, N+1])
M_t[:,0] = 1
R[:,0]=f(0,0)
t = np.zeros([N+1])
dt = T / float(N)


for i in range(0,N):
    W[:,i+1] = W[:, i] + Z[:,i]*sqrt(dt)
    dW = W[:,i+1] - W[:,i]
    R[:, i+1] =  R[:, i] +  (θ(t[i]) - λ*R[:, i] )*dt + σ*dW 
    M_t[:, i+1] = M_t[:,i] * np.exp( (R[:, i+1] + R[:, i])/2.0 * dt)
    t[i+1] = t[i]+ dt


fig = plt.figure(figsize=(14,8)) # create a figure where we will put our plots
plt.plot(t, np.transpose(R[0:100,:]), linestyle='-')
plt.grid()
plt.xlabel("time")
plt.ylabel("r(t)");
plt.title('HW1F Short Rate Simulation', fontsize = 24);

Value = zeros_like(R)
E = zeros_like(R)
EE = np.zeros([N+1])
PFE = np.zeros([N+1])
PFE2 = np.zeros([N+1])

for j in range(N):
    V = IRS_HW(t=t[j], T = arange(2,11,1).reshape(9,1) , τ = 1, N = 100, K = 0.0293,a = λ, σ = σ, r_t= R[:,j], irs_type='R')
    Value[:,j] = np.nan_to_num(V)  
    E[:,j] = np.maximum(V,0)
    EE[j] = np.mean(E[:,j] / M_t[:,j] )
    PFE[j] = np.quantile(E[:,j], .95)
    PFE2[j] = np.quantile(E[:,j], .99)
        
     
fig = plt.figure(figsize=(14,8)) # create a figure where we will put our plots
plt.plot(t, np.transpose(Value[0:100,:]), linestyle='-')
plt.grid()
plt.xlabel("time")
plt.ylabel("V(t)");
plt.title('Swap Value V(t)', fontsize = 24);


fig = plt.figure(figsize=(14,8)) # create a figure where we will put our plots
plt.plot(t, np.transpose(E[0:100,:]), linestyle='-')
plt.grid()
plt.xlabel("time")
plt.ylabel("E(t)");
plt.title('Positive Exposure E(t)', fontsize = 24);


fig = plt.figure(figsize=(14,8)) # create a figure where we will put our plots
ax = fig.add_subplot(1,1,1)
ax.plot(t, np.transpose(EE), linestyle='-', label = 'EE(t)')
ax.plot(t, np.transpose(PFE), linestyle='-' , label = 'PFE(t) 95%' )
ax.plot(t, np.transpose(PFE2), linestyle='-' , label = 'PFE(t) 99%' )
plt.grid()
plt.xlabel("time")
ax.legend(fontsize = 'large')
ax.set_title('Exposure Plots', fontsize = 24);
```
